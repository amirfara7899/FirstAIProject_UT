```python
df.loc[df['FUELTYPE'] == 'D', "FUELTYPE"] = '0'# this method change the value cell of column FUELTYPE where is X to 0
df.loc[df['FUELTYPE'] == 'E', "FUELTYPE"] = '1'
df.loc[df.FUELTYPE == "X","FUELTYPE"] = '2'
df.loc[df['FUELTYPE'] == "Z","FUELTYPE"] = '3'
# df['FUELTYPE'] = df['FUELTYPE'].fillna('5')# this method fill the NAN Cell of column FUELTYPE to 5
# df.astype({'FUELTYPE': 'int32'}).dtypes # this method change the type of column to int32
```

```python
df['MODEL'] = df['MODEL'].fillna(df['MODEL'].mode()[0])#find mode of MODEL and fill the NAN cells with it
df['ENGINESIZE'] = df['ENGINESIZE'].fillna(df['ENGINESIZE'].mean())#find mean of ENGINESIZE and fill the NAN cells with it
df['CYLINDERS'] = df['CYLINDERS'].fillna(df['CYLINDERS'].mean())
#pd.set_option('display.max_rows', 10)
print(df[['MODEL','CYLINDERS', 'ENGINESIZE']])
```

```python
filtered_df = df.loc[(pd.isna(df['CO2EMISSIONS']))]# method of isna check the varaible(cell) is Nan or not
filtered_df
```

```python
%%time
ave_less_240 = df['CO2EMISSIONS'][df['CO2EMISSIONS'] < 240].mean()#mean for cell of CO2EMISSIONS that less than 240 
ave_More_300 = df['CO2EMISSIONS'][df['CO2EMISSIONS'] > 300].mean()#mean for cell of CO2EMISSIONS that more than 300
print(ave_less_240)
print(ave_More_300)
```

```python
%%time
number_1 = 0
number_2 = 0
total_1 = 0
total_2 = 0
for name, row in df['CO2EMISSIONS'].iteritems():
    if row < 240:
        total_1 += row
        number_1 += 1
for name, row in df['CO2EMISSIONS'].iteritems():
    if row > 300:
        total_2 += row
        number_2 += 1
print(total_1/number_1)
print(total_2/number_2)
```
